```python
#import the library
import json, requests

#url for the boredAPI we are using, boredAPI does not use a key to access their API
base_url = "https://bored.api.lewagon.com/api/activity"

#print a welcome message for fun! 
print("Welcome to the Bored API!")

#create a variable for when we request the URL using request.get.
response = requests.get(base_url)

#create a variable for the response.json text
unformatted_data = response.json()

#from the unformatted data create a variable for the various data we receive
activity_data = unformatted_data["activity"]

#from the unformatted data create a variable for the various data we receive
type_data = unformatted_data["type"]

#from the unformatted data create a variable for the various data we receive
participants_data = unformatted_data["participants"]

#from the unformatted data create a variable for the various data we receive
link_data = unformatted_data["link"]

#from the unformatted data create a variable for the various data we receive
key_data = unformatted_data["key"]

#from the unformatted data create a variable for the various data we receive
accessibility_data = unformatted_data["accessibility"]

#from the unformatted data create a variable for the various data we receive
price_data = unformatted_data["price"]


#print a space
print ()

#print the data
print(f"Want to '{activity_data}'?")

#print a space
print ()

#print the data
print(f"This is a '{type_data}' activity.")

#print a space
print ()

#print the data
print(f"It takes '{participants_data}' people/persons to do this activity.")

#print a space
print ()

#print the data
print(f"Here's a link to learn more: '{link_data}'!")

#print a space
print ()

#print the data
print(f"This is information that can be found from the bored.api website (this number will lead you back to the '{activity_data}' key): '{key_data}'")

#print a space
print ()

#print the data
print(f"Cost associated with this activity (0 to 1, 0 is free): '{price_data}'")

#print a space
print ()

#print the data
print(f"Level of accessibility from 0 to 1 (0 being the most accessible): '{accessibility_data}'")
```
